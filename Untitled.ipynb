```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
import seaborn as sns
```

```python
np.random.seed(0)
m=100
x = np.linspace(0,10,m).reshape(m,1)
y = x**2 + np.random.randn(m,1)
```

```python
_=plt.scatter(x,y)
```

```python
model = SVR(C=100)
model.fit(x,y)
model.score(x,y)
```

```python
prediction=model.predict(x)
```

```python
_=plt.scatter(x,y)#pour le nuage des points
_=plt.plot(x , prediction, c = 'r') # pour la ligne de regression
```
